```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1

```python
df = pd.read_csv("data/usa_deaths.csv",index_col='date',parse_dates=['date'])
df
```

```python
deces_total = df.sum()
deces_total
```

# 5

```python
df.sum().sum()
```

# 6

```python
x1 = df['Texas'].cumsum()
x2 = df['California'].cumsum()
plt.figure(figsize=(10,6))
sns.lineplot(x='date',y=x1,data=df,label='Morts Texas')
sns.lineplot(x='date',y=x2,data=df,label='Morts California')
plt.legend()
plt.title('Comparaidon des décès')
plt.show()
```

# 7

```python
df.mean().sort_values(ascending=False)
```

# 8

```python
plt.figure(figsize=(14,8))
sns.barplot(x=deces_total.index,y = deces_total.values)
plt.title("Décès total par etat")
plt.xticks(rotation=90)
plt.xlabel('Etat')
plt.ylabel('Nombre total de décès')
plt.show()
```

# 9

```python
etats = ['Texas','California','New York']
plt.figure(figsize=(12, 8))
newDf = df[etats]
sns.boxplot(data=newDf,palette='Set1')
plt.xlabel('Etats')
plt.ylabel('Deces')
plt.show()
```
