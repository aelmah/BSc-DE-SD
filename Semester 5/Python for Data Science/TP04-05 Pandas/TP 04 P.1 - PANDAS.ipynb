# Q8

```python
countries = ['Morocco','Brazil','Turkey','Albania']
for country in countries:
    # df[country].plot(legend = country)
    plt.plot(df[country],label = country)
plt.legend()
plt.xlabel('Periode')
plt.ylabel('Décès')
plt.title('Décès dans les quatre pays')
plt.show()
```

# Q9

```python
Mdf = df['Morocco']
Mdf.resample('M').sum() #'M' month ; 'D' days ; 'W' weekly
```

```python
Mdf.resample('M').sum().plot(legend='Morocco')
plt.xlabel('Periode')
plt.ylabel('Décès')
plt.title('Décès chaque mois')
plt.show()
```
