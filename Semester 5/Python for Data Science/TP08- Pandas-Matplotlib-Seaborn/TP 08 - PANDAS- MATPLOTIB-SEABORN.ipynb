# 7

```python
df['DATE.OCC'] = pd.to_datetime(df['DATE.OCC'])
df['Year'] = df['DATE.OCC'].dt.year
crimeParAns = df['Year'].value_counts()
crimeParAns = crimeParAns.sort_index()
sns.lineplot(crimeParAns)
plt.show()
```

# 8

```python
top_10_crimes = df['CrmCd.Desc'].value_counts().nlargest(10)
sns.barplot(x = top_10_crimes.index,y = top_10_crimes.values,palette='Set1',hue=top_10_crimes.index)
plt.xticks(rotation=90)
plt.show()
```

# 9

```python
top_10_AreaName = df['AREA.NAME'].value_counts().head(10)
sns.barplot(x = top_10_AreaName.index,y = top_10_AreaName.values,palette='Set2',hue=top_10_AreaName.index)
plt.xticks(rotation=90)
plt.show()
```

# 11

```python
df['DATE.OCC'] = pd.to_datetime(df['DATE.OCC'])
crimes_southwest = df[(df['DATE.OCC'].dt.year >= 2012) 
                      & (df['DATE.OCC'].dt.year <= 2015) 
                      & (df['AREA.NAME'] == 'Southwest')]
print("total des crimes : ",len(crimes_southwest))
```

# 12

```python
crimes_southwest = df[df['AREA.NAME'] == 'Southwest']
top_30_crimes_southwest = crimes_southwest['CrmCd.Desc'].value_counts().head(30)
plt.figure(figsize=(12,6))
sns.barplot(top_30_crimes_southwest)
plt.xticks(rotation=45,ha='right')
plt.xlabel("Crimes")
plt.ylabel("Nombre de crimes")
plt.title("30 principeaux crimes dans 'Southwest'")
plt.show()
```

# 13

```python
crime_count_2015_2016 = df[(df['DATE.OCC'].dt.year >= 2015) & (df['DATE.OCC'].dt.year <= 2016)]
sns.countplot(x = crime_count_2015_2016['DATE.OCC'].dt.year)
plt.title("Numbre d'incidents signales entre 2015 et 2016")
plt.xlabel('Year')
plt.ylabel("Numbre d'incidents")
plt.show()
```
