```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1

```python
df1 = pd.read_excel('data/filmes.xls',sheet_name=0,index_col=0)
df2 = pd.read_excel('data/filmes.xls',sheet_name=1,index_col=0)
df3 = pd.read_excel('data/filmes.xls',sheet_name=2,index_col=0)
```

# 2

```python
df = pd.concat([df1,df2,df3])
```

# 3

```python
df1 = df1.dropna()
df2 = df2.dropna()
df3 = df3.dropna()
```

# 4

```python
df1.drop_duplicates()
df2.drop_duplicates()
df3.drop_duplicates()
df = pd.concat([df1,df2,df3])
df
```

# 5

methode1


```python
plt.figure(figsize=(12,6))
sns.countplot(x='Language',data=df,palette='Set1',hue='Language') 
plt.xticks(rotation=90)
plt.xlabel('Langue')
plt.ylabel('Nombre de filme')
plt.title('Nombre de fime selon la langue')
plt.show()
```

methode2

```python
plt.figure(figsize=(12,6))
langue_count = df['Language'].value_counts().reset_index()
sns.barplot(x='Language',y='count',data=langue_count,palette='Set1',hue='Language')
plt.xticks(rotation=90)
plt.xlabel('Langue')
plt.ylabel('Nombre de filme')
plt.title('Nombre de fime selon la langue')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df1['IMDB Score'],label='Anne 1900',kde=True,color='blue')
sns.histplot(df2['IMDB Score'],label='Anne 2000',kde=True,color='green')
sns.histplot(df3['IMDB Score'],label='Anne 2010',kde=True,color='red')
plt.legend()
plt.xlabel("Notes")
plt.ylabel("Frequence")
plt.show()
```

# 7

```python
plt.figure(figsize=(12,6))
budgetDf1 = df1.groupby('Year')['Budget'].mean()
budgetDf2 = df2.groupby('Year')['Budget'].mean()
budgetDf3 = df3.groupby('Year')['Budget'].mean()
sns.lineplot(budgetDf1,label='Anne 1900',marker='o')
sns.lineplot(budgetDf2,label='Anne 2000',marker='*')
sns.lineplot(budgetDf3,label='Anne 2010',marker='+')
plt.title('Evolution des budgets moyennes des filmes par decennies')
plt.xlabel('Année')
plt.ylabel("Budget moyen")
plt.show()
```

# 8

```python
top_genre_df1 = df1['Genres'].value_counts().head()
top_genre_df2 = df2['Genres'].value_counts().head()
top_genre_df3 = df3['Genres'].value_counts().head()
plt.figure(figsize=(15,5))
plt.subplot(131)
sns.barplot(top_genre_df1,color="red")
plt.title("Genre plus frequents (Année 1900)")
plt.xticks(rotation=-90)
plt.subplot(132)
sns.barplot(top_genre_df2,color="green")
plt.title("Genre plus frequents (Année 2000)")
plt.xticks(rotation=-90)
plt.subplot(133)
sns.barplot(top_genre_df3,color="blue")
plt.title("Genre plus frequents (Année 2010)")
plt.xticks(rotation=-90)
plt.show()
```
