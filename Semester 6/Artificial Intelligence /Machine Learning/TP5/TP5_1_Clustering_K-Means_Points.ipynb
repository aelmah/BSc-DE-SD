#### TP5: Clustering K-Means

##### * Dataset: Points

###### Packages 

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.metrics import silhouette_score
import numpy as np
```

#### K-means 

```python
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv('Dataset/points.csv')
df
```

##### * Visualisation des données 

```python
plt.scatter(df['x'] , df['y'])
plt.title('Visualisation des données Points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Répartition des données

```python
X= df[['x','y']]
```

##### Création du modèle K-Means

```python
Model=KMeans(n_clusters=4)
```

##### Ajustement du modèle

```python
Model.fit(X)
```

##### Détermination des clusters

```python
cluster_labels = Model.labels_
```

```python
plt.scatter(df['x'], df['y'], c= df['cluster'], cmap= 'viridis')
plt.title('Visualisation des données après le clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

##### Évaluation du modèle

* INFO:
 en utilise la métrique de silhouette pour évaluer la qualité du clustering. Un score de silhouette proche de 1 indique un clustering bien séparé.

```python
silhouette_avg = silhouette_score(X, cluster_labels)
print(f"Silhouette Score : {silhouette_avg}")
```
