#### TP5: Clustering BDSCAN avec normalisation

* Dataset : Points

##### Packages 

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('Dataset/points.csv')
df
```

##### Visualisation des données

```python
plt.scatter(df['x'] , df['y'])
plt.title('Visualisation des données Points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##### Répartition des données

```python
X= df[['x','y']]
```

##### Normalisation 

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

##### Création du modèle 

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps= 0.5,min_samples= 5)
```

##### Ajustement du modèle 

```python
dbscan_labels = model.fit_predict(X_normalized)
```

##### Détermination des clusters

```python
cluster_labels = model.fit_predict(X)
```

##### Visualisation 

```python
plt.scatter(X_normalized[:, 0], X_normalized[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('Données après la phase de clustering avec DBSCAN (normalisées)')
plt.xlabel('X_normalized')
plt.ylabel('Y_normalized')
plt.show()
```
